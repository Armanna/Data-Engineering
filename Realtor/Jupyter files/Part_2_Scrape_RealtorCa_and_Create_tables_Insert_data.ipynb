```python
def url_options(current_url):
    dict = {}    
    
    zoom_level = current_url.find('ZoomLevel')
    if current_url[zoom_level + 11] == '&':
        ZoomLevel = current_url[zoom_level + 10]
    else:
        ZoomLevel = current_url[zoom_level + 10:zoom_level + 12]
    
    dict['ZoomLevel'] = ZoomLevel
        
    latitude_max = current_url.find('LatitudeMax')
    LatitudeMax = current_url[latitude_max + 12:latitude_max + 20]
    
    dict['LatitudeMax'] = LatitudeMax
    
    longitude_max = current_url.find('LongitudeMax')
    LongitudeMax = current_url[longitude_max + 13:longitude_max + 22]
    
    dict['LongitudeMax'] = LongitudeMax
    
    latitude_min = current_url.find('LatitudeMin')
    LatitudeMin = current_url[latitude_min + 12:latitude_min + 20]
    
    dict['LatitudeMin'] = LatitudeMin

    
    longitude_min = current_url.find('LongitudeMin')
    LongitudeMin = current_url[longitude_min + 13:longitude_min + 22]
    
    dict['LongitudeMin'] = LongitudeMin
    
    return dict
```

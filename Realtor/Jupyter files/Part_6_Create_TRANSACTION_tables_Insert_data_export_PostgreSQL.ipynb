## We have 4 transaction tables
- 1) Estate_Broker_relationship
- 2) Broker_behavior
- 3) Contract
- 4) Estates

### In the first, in the "Estate_Broker_relationship" table estates and brokers are in "in touch" relationship.
### Then, with the package "time", after avery 10 seconds the result status of about every 20000 estates would be changed from "in touch" status to "success" or "failed".
### If the result_status is changed to "success", in the "broker_behavior" table the "estate_success" cell of a corresponding broker would be added by 1, otherwise, if the result_status is changed to "failed", the "estate_failed" cell would be added by 1.
### In parralel, with the "success" result the "Contract" table will be added by 1 row.
### In parralel, the status of each estate in "Estate" table would be changed from "is selling" to "sold".
### The process will continue till all the estates would be sold.
### So let's start the engine.

```python
def Enjoy(Estate_Broker_relationship,Broker_behavior,Contract,Estates):
    
    n = len(Estate_Broker_relationship) // 5
    m = len(Estate_Broker_relationship) - 4 * n
    
    estates = random.sample(range(1, len(Estates) + 1), len(Estates))
    estates_1 = estates[:n]
    estates_2 = estates[n:2 * n]
    estates_3 = estates[2 * n:3 * n]
    estates_4 = estates[3 * n:4 * n]
    estates_5 = estates[4 * n:]
    
    est = [estates_1,estates_2,estates_3,estates_4,estates_5]
    
    buyers = []
    for i in range(len(Estates)):
        buyers.append(random.randint(1,90001))
    b = -1    
    
    WORK = True  
    
    contract_id = 0
    
    while WORK:
        if len(Estates[Estates['status'] == 1]) == 0:
            WORK = False
            
        for i in est:
            s = len(i) // 1000
            f = len(i) - s * 1000
            success = i[:s]
            fail = i[s:]
            
            for j in success:
                index_1 = Estate_Broker_relationship['estate_id'][Estate_Broker_relationship['estate_id'] == j].index[0]
                Estate_Broker_relationship['result_id'][index_1] = 2
                
                index_2 = Estate_Broker_relationship['broker_id'][index_1] - 1
                Broker_behavior['estates_successed'][index_2] += 1
                
                contract_id += 1
                Contract['contract_id'] = contract_id
                estate_id = j
                broker_id = index_2 + 1
                b += 1
                buyer_id = buyers[b]
                owner_id = Estates['owner_id'][j - 1]
                date = fake.date_between_dates(date_start=datetime(2020,1,1), date_end=datetime.now())
                price = Estates['price'][j - 1]
                Contract = Contract.append({'contract_id':contract_id,
                                           'estate_id':estate_id,
                                           'broker_id':broker_id,
                                           'buyer_id':buyer_id,
                                           'owner_id':owner_id,
                                           'date':date,
                                           'price':price},ignore_index=True)
                
                Estates['status'][j - 1] = 2
                
            for j in fail:
                index_1 = Estate_Broker_relationship['estate_id'][Estate_Broker_relationship['estate_id'] == j].index[0]
                Estate_Broker_relationship['result_id'][index_1] = 3
                index_2 = Estate_Broker_relationship['broker_id'][index_1] - 1
                Broker_behavior['estates_failed'][index_2] += 1
                
                failed_broker = Estate_Broker_relationship['broker_id'][index_1]
                br = list(range(1,failed_broker)) + list(range(failed_broker + 1, len(Broker_behavior) + 1))
                Estate_Broker_relationship['broker_id'][index_1] = random.choice(br)


            for j in fail:
                index_1 = Estate_Broker_relationship['estate_id'][Estate_Broker_relationship['estate_id'] == j].index[0]
                Estate_Broker_relationship['result_id'][index_1] = 2
                
                index_2 = Estate_Broker_relationship['broker_id'][index_1] - 1
                Broker_behavior['estates_successed'][index_2] += 1
                
                contract_id += 1
                Contract['contract_id'] = contract_id
                estate_id = j
                broker_id = index_2 + 1
                b += 1
                buyer_id = buyers[b]
                owner_id = Estates['owner_id'][j - 1]
                date = fake.date_between_dates(date_start=datetime(2020,1,1), date_end=datetime.now())
                price = Estates['price'][j - 1]
                Contract = Contract.append({'contract_id':contract_id,
                                           'estate_id':estate_id,
                                           'broker_id':broker_id,
                                           'buyer_id':buyer_id,
                                           'owner_id':owner_id,
                                           'date':date,
                                           'price':price},ignore_index=True)
                
                Estates['status'][j - 1] = 2
                
                
#             print(Estate_Broker_relationship)
#             print(Broker_behavior)
#             print(Contract)
#             print(Estates)
            
            sleep(10)   # It may be sleepd for a week, for example
            return (Estate_Broker_relationship,Broker_behavior,Contract,Estates)
              
        
        
        
```
